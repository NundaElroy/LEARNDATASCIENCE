```python
import pandas as pd

df = pd.read_csv("data_recess.csv")

print(df.isnull().sum())
```

```python
#checking the data types
print(df.dtypes)
```

```python
#filling in empty columns 
for col in ['Duration','Pulse','Maxpulse','Calories']:
     df[col].fillna(df[col].mean(),inplace= True)

print(df)
   
```

```python
#summmary stats
print("Summary statistics")

print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

# Histogram for duration
# Set up the figure size
plt.figure(figsize=(15, 10))

# duration histogram 
plt.subplot(2, 2, 2)
sb.histplot(df['Duration'], bins=30, kde=True)
plt.title('Duration Distribution')
# pulse histogram
plt.subplot(2, 2, 1)
sb.histplot(df['Pulse'], bins=30, kde=True)
plt.title('Pulse Distribution')

# Exercise Plot for Maxpulse and Colories for Histrogram
```

```python
##Exercise
# calories histogram 
plt.subplot(2, 2, 2)
sb.histplot(df['Calories'], bins=30, kde=True)
plt.title('Calories Distribution')
# maxpulse histogram
plt.subplot(2, 2, 1)
sb.histplot(df['Maxpulse'], bins=30, kde=True)
plt.title('Maxpulse Distribution')
```

```python
# Scatter plot of 'Duration' vs 'Pulse'
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sb.scatterplot(x='Duration', y='Pulse', data=df)
plt.title('Duration vs Pulse')
# Scatter plot of 'Pulse'
plt.subplot(2, 2, 2)
sb.scatterplot(x='Pulse', y='Calories', data=df)
plt.title('Pulse vs Calories')
```

```python
# Correlation heatmap
corr = df.corr()
sb.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
